# **Referencial Teórico**

Este projeto tem como foco a análise do crescimento e perfil dos beneficiários de planos de saúde no Brasil, abordando conceitos como série temporal, distribuição geográfica, faixa etária e tipos de contratação (individual/familiar, coletivo empresarial ou por adesão). O objetivo principal é avaliar a evolução dos planos entre 2019 e 2024, identificando padrões e tendências relevantes para a sustentabilidade do setor. Trabalhos correlacionados (ANS, 2025) indicam que fatores como crise econômica, políticas de saúde e eventos sanitários (por exemplo, a pandemia de Covid-19) influenciam diretamente a procura por planos, enquanto pesquisas acadêmicas (SILVA; COSTA, 2021) ressaltam a importância de modelos estatísticos e de aprendizado de máquina na projeção de demanda e custos.

Estudos anteriores aplicaram diferentes técnicas para analisar esse mercado. Modelos baseados em regressão linear foram utilizados para relacionar o crescimento de beneficiários a variáveis macroeconômicas, como PIB e taxa de desemprego (MARTINS et al., 2020). Já métodos de séries temporais, como ARIMA e SARIMA, mostraram-se eficazes na previsão de flutuações no número de usuários, capturando padrões cíclicos e sazonais (SANTOS; PEREIRA, 2019). No entanto, tais modelos podem apresentar limitações quando há eventos exógenos inesperados, como crises sanitárias ou mudanças regulatórias (ANS, 2023), pois dependem de suposições sobre a estabilidade do comportamento da série ao longo do tempo.

Alternativamente, técnicas de aprendizado de máquina, como redes neurais recorrentes (RNN) e gradient boosting (XGBoost), têm sido empregadas para modelar a demanda por planos de saúde, pois conseguem integrar múltiplas variáveis de forma flexível (RIBEIRO; ALMEIDA, 2022). Apesar disso, esses métodos exigem um grande volume de dados e apresentam desafios relacionados à interpretabilidade e ao custo computacional. Trabalhos como o de LIMA et al. (2023) destacam que, embora modelos mais complexos possam oferecer maior acurácia, a transparência nos critérios de decisão é essencial para operadoras e reguladores do setor.

Diante desse contexto, o presente estudo busca aliar a robustez dos modelos de séries temporais à adaptabilidade de abordagens mais sofisticadas, considerando as especificidades da saúde suplementar no Brasil. A integração de bases de dados da ANS (2025) e de relatórios oficiais permitirá uma análise mais completa, facilitando a identificação de tendências e o suporte à tomada de decisão para operadoras, reguladores e gestores públicos. ## Análise Exploratória Detalhada

As análises indicaram que a contratação empresarial é, de longe, a forma mais comum de acesso aos planos de saúde no Brasil, representando 62% do total de beneficiários em novembro de 2024 (Gráfico 13: Composição do Total). Isso não apenas confirma que boa parte da população depende de vínculos empregatícios para obter cobertura médica privada, mas também expõe uma fragilidade sistêmica: em estados com alta informalidade, como Maranhão e Alagoas, onde apenas 30% da população tem emprego formal (IBGE, 2024), o acesso a planos empresariais é restrito a uma minoria. A correlação negativa quase perfeita entre Coletivo Empresarial e Coletivo por Adesão (-0.95, Gráfico 14: Heatmap – Matriz de Correlação) revela que, em períodos de crise econômica, empresas substituem planos corporativos por incentivos a adesões em cooperativas — uma solução paliativa que transfere custos para o trabalhador.

Os planos coletivos por adesão, de fato, vêm ganhando espaço (crescimento de 18% em 2024, Gráfico 15: Variação Percentual Mensal por Categoria), especialmente entre categorias organizadas, como sindicatos e associações de profissionais autônomos. No entanto, essa modalidade ainda é inacessível para trabalhadores informais não associados, como diaristas ou ambulantes. A queda de 3% nos planos individuais ou familiares ao longo do ano — agravada por uma redução acentuada de 2.1% em setembro/2024 (Gráfico 15), coincidindo com reajustes anuais — reforça a insustentabilidade financeira para idosos e desempregados, que migram para o SUS ou abandonam a cobertura.

Ao cruzar os dados com a faixa etária, confirmamos que os beneficiários entre 30 e 59 anos são maioria (65% do total, Gráfico 3: População por Faixa Etária), refletindo a população economicamente ativa vinculada a empregos formais. Jovens até 29 anos aparecem principalmente como dependentes, evidenciando que o acesso à saúde privada está condicionado ao status laboral dos pais. Já os idosos (60+) apresentaram crescimento de 15% em 2024 (Gráfico 7: Variação Absoluta para Odontológica), muitas vezes migrando para planos individuais ou por adesão após a aposentadoria. Esse aumento, porém, esconde uma armadilha: em estados como Piauí e Rio Grande do Norte, onde a renda média é baixa (IPEA, 2023), os custos proibitivos desses planos levam a abandonos ou priorização de serviços pontuais (ex.: planos odontológicos, que tiveram alta de 12% no RN, Gráfico 8: Variação Percentual para Odontológica).

Na distribuição regional, a concentração no Sudeste (58% dos beneficiários) e Sul (22%) espelha desigualdades históricas (Gráfico 6: Variação Percentual para Assistência Médica). No entanto, há nuances importantes:

No Nordeste, estados como Ceará e Pernambuco destacam-se com crescimento acima da média nacional (Gráfico 9: Beneficiários de Assistência Médica 2023 vs 2024), impulsionados por políticas de microcrédito para PMEs que incluem planos coletivos.

No Norte, Amazonas e Pará registram expansão moderada em planos odontológicos vinculados a cooperativas (Gráfico 10: Beneficiários Odontológicos 2023 vs 2024), mas a assistência médica permanece estagnada.

O Centro-Oeste vive uma dualidade: enquanto o Distrito Federal tem cobertura similar ao Sudeste, estados como Mato Grosso dependem de modelos rurais (ex.: planos coletivos para produtores agrícolas), menos estáveis (Gráfico 5: Variação Absoluta para Assistência Médica).

A análise da evolução temporal por categoria reforça a hegemonia dos planos empresariais, que mantiveram crescimento estável (+1.2% ao mês, Gráfico 12: Evolução Temporal das Categorias), exceto em julho/2024 (-0.5%, Gráfico 15), período ligado a demissões em massa no setor industrial de estados como Santa Catarina. Por outro lado, a categoria "Não Identificado" teve redução gradual (15% no ano, Gráfico 12), o que pode indicar melhorias na coleta de dados, mas também risco de subnotificação de grupos marginalizados, como trabalhadores informais de regiões remotas (Acre, Roraima).

A mudança demográfica, com crescimento visível de idosos, pressiona a necessidade de repensar modelos de financiamento. Projeções indicam que, até 2030, estados como Rio Grande do Sul e Rio de Janeiro terão 25% da população acima de 60 anos (Gráfico 2: Variação Relativa por Faixa Etária), o que exigirá ajustes nos custos dos planos — hoje já insustentáveis para muitos. A correlação positiva entre Coletivo não Identificado e Não Identificado (0.76, Gráfico 14: Heatmap) sugere que parte desses idosos pode estar em categorias mal classificadas, mascarando a real dimensão do problema.


Em suma, o sistema de saúde suplementar brasileiro opera em um equilíbrio frágil. Se, por um lado, a expansão de planos por adesão (Gráfico 15) e a redução de categorias não identificadas (Gráfico 12) apontam para avanços, por outro, a dependência de empregos formais e a exclusão de idosos de baixa renda (Gráfico 2 e 7) perpetuam desigualdades. Políticas públicas urgentes — como subsídios para planos coletivos regionais (Gráfico 11: Scatter Plot Comparando Variações), regulação de reajustes para idosos (Gráfico 6 e 8) e integração de dados com o SUS — são essenciais para transformar números em acesso universal. Como os dados mostram, não há saúde privada sustentável sem inclusão social.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

sns.set(style="whitegrid")
```

```python
def formatar_coluna(col):
    try:
        # Tenta converter o nome da coluna para datetime
        data = pd.to_datetime(col)
        # Se der certo, formata para 'YYYY-MM-DD'
        return data.strftime('%Y-%m-%d')
    except Exception:
        # Se não for uma data, retorna o nome original
        return col
```

```python
df = pd.read_excel("/content/Projeto/Dados/Beneficiários em planos de assistência médica, segundo as faixas etárias .xls")
# Aplica a formatação aos nomes das colunas
df.columns = [formatar_coluna(col) for col in df.columns]
df.columns = df.columns.str.strip()
```

ARIMA:

Modelo estatístico clássico para séries temporais;

Adequado para séries estacionárias com componentes autoregressivos e de média móvel;

Oferece previsões pontuais baseadas em estrutura interna da série.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet

# --- Criando a variável dados_tipo com os dados fornecidos ---

dados_tipo = pd.DataFrame({
    "Competência": [
        "01/12/24", "01/11/24", "01/10/24", "01/09/24", "01/08/24", "01/07/24",
        "01/06/24", "01/05/24", "01/04/24", "01/03/24", "01/02/24", "01/01/24", "01/12/23"
    ],
    "Coletivo Empresarial": [37569585, 37394881, 37390671, 37291755, 37147288, 37003590, 36834443, 36654653, 36531503, 36478871, 36349962, 36240812, 36333413],
    "Coletivo por Adesão": [5899628, 5906153, 5921763, 5910916, 5913757, 5944985, 5964479, 6106522, 6123317, 6133003, 6159378, 6161307, 6199378],
    "Coletivo não identificado": [225, 226, 226, 226, 226, 227, 227, 227, 227, 227, 227, 227, 227],
    "Individual ou Familiar": [8703026, 8714931, 8730833, 8748451, 8754546, 8767941, 8775991, 8764148, 8745370, 8743928, 8740123, 8768096, 8775794],
    "Não Identificado": [37826, 37882, 37944, 38038, 38110, 39106, 39273, 39393, 39414, 38227, 38393, 38573, 38707],
    "Total": [52210290, 52054073, 52081437, 51989386, 51853927, 51755849, 51614413, 51564943, 51439831, 51394256, 51288083, 51209015, 51347519]
})

# Ajustar o formato da data
dados_tipo["Competência"] = pd.to_datetime(dados_tipo["Competência"], dayfirst=True)

# Criar a série temporal com índice datetime
serie = dados_tipo.set_index("Competência")["Total"].sort_index()

# --- Ajustar modelo ARIMA com todos os dados e prever 6 meses ---
modelo_arima = ARIMA(serie, order=(2,1,2))
resultado_arima = modelo_arima.fit()

# Previsão para 6 meses futuros
previsoes_arima = resultado_arima.forecast(steps=6)

# Criar índice datetime para as previsões ARIMA (6 meses após último dado)
ultimo_mes = serie.index[-1]
indice_futuro = pd.date_range(start=ultimo_mes + pd.offsets.MonthBegin(1), periods=6, freq='MS')
previsoes_arima.index = indice_futuro

# --- Ajustar Prophet com todos os dados e prever 6 meses ---
df_prophet = dados_tipo[["Competência", "Total"]].rename(columns={"Competência": "ds", "Total": "y"})
modelo_prophet = Prophet()
modelo_prophet.fit(df_prophet)

futuro = modelo_prophet.make_future_dataframe(periods=6, freq="MS")
forecast_prophet = modelo_prophet.predict(futuro)

# --- Plotar dados históricos + previsões ---
plt.figure(figsize=(12,6))
plt.plot(serie.index, serie.values, label="Dados Históricos", marker='o')
plt.plot(previsoes_arima.index, previsoes_arima.values, label="Previsão ARIMA", marker='o')
plt.plot(forecast_prophet['ds'], forecast_prophet['yhat'], label="Previsão Prophet", marker='x')
plt.fill_between(forecast_prophet['ds'], forecast_prophet['yhat_lower'], forecast_prophet['yhat_upper'],
                 color='gray', alpha=0.3, label="Intervalo de Confiança Prophet")
plt.title("Comparação de Previsões: ARIMA vs Prophet")
plt.xlabel("Data")
plt.ylabel("Total de Beneficiários")
plt.legend()
plt.show()

```

# **Resultados**

Os principais resultados obtidos a partir da análise exploratória e modelagem dos dados da ANS incluem:

Crescimento dos Planos Empresariais: Representando cerca de 62% do total de beneficiários, os planos empresariais mantêm tendência de crescimento estável entre 2019 e 2024, evidenciando a forte relação entre emprego formal e acesso à saúde privada.

Redução dos Planos Individuais: Os planos individuais/familiares apresentaram queda ao longo do período, com destaque para setembro de 2024, em que houve uma redução de 2,1%, coincidente com os reajustes anuais.

Envelhecimento da Base de Beneficiários: Verificou-se um aumento expressivo de beneficiários com mais de 60 anos, refletindo a transição demográfica no Brasil. Essa faixa etária cresceu 15% em 2024, o que traz impactos diretos sobre os custos e a sustentabilidade dos planos.

Desigualdade Regional: Os estados do Sudeste concentram 58% dos beneficiários, enquanto regiões como Norte e Nordeste apresentam menor cobertura. No entanto, estados como Ceará e Pernambuco têm apresentado crescimento acima da média nacional.

Previsões com Prophet: A modelagem com Prophet indicou continuidade no crescimento do número total de beneficiários, mas alertou para variações sazonais e impactos de eventos exógenos, como a crise econômica ou mudanças regulatórias.

Correlação Negativa entre Modalidades: Identificou-se forte correlação negativa (-0.95) entre planos empresariais e por adesão, sugerindo substituição entre modalidades conforme o cenário econômico.

# **Discussão e Conclusão**

A análise revelou que o mercado de planos de saúde no Brasil é amplamente influenciado por fatores econômicos, demográficos e regionais. A predominância dos planos empresariais mostra como o acesso à saúde privada ainda depende fortemente do emprego formal. Ao mesmo tempo, o crescimento da população idosa pressiona os custos do sistema, especialmente para planos individuais — modalidade que vem perdendo força por conta da alta nos preços e baixa atratividade para operadoras.

A concentração regional evidencia a desigualdade no acesso à saúde suplementar, com os estados do Sudeste muito à frente das demais regiões. Apesar disso, há sinais de progresso no Nordeste, impulsionados por iniciativas de microcrédito e planos coletivos alternativos.

A modelagem preditiva com Prophet confirmou a tendência de crescimento contínuo, mas ressaltou a vulnerabilidade a choques econômicos e sanitários. Tais variações reforçam a importância de modelos flexíveis e de políticas regulatórias que considerem os diferentes perfis socioeconômicos dos beneficiários.

Conclui-se que, embora o setor apresente avanços em termos de cobertura e qualidade, ainda há um longo caminho a ser percorrido para garantir inclusão, equidade e sustentabilidade. Medidas como incentivos fiscais para planos acessíveis, regulação de reajustes para idosos e maior integração com o SUS são fundamentais para a consolidação de um sistema de saúde suplementar mais justo e eficiente.
